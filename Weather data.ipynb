```python
import seaborn as sb
import pandas as pd

dataset = pd.read_csv("weather_data.csv")
dataset = dataset.rename(columns = {"01/07/2014" : "The date", "76" : "Actual average °F that day", "66" : "Actual minimum °F that day",
                                   "86" : "The actual maximum °F that day", "66.1" : "The average minimum °F of that day since 1901", "85" : "The average max °F that day since 1901",
                                    "48" : "The lowest °F that day since 1901", "97" : "The highest °F that day since 1901", "0.13" : "The actual amount of rain that day", 
                                    "0.16" : "The average amount of rain on that day since 1901", "5.09" : "The highest amount of rain on that day since 1901"})

sb.heatmap(dataset.corr(), annot = True) #For getting the numbers on here we should add annot = True, 
# by this heatmap visualisation we can see that rain itself isn't common whether it is because of most of the right side (Last three columns/about rain) 
#that there is a low correlation with other types of non rain columns 

#Thus under my own assumption, I conclude that with this data specifies a place in australia that isn't too hot or too cold but doesn't receive rainfall

#I chose the heatmap for a quicker understanding in the data with the darker colour symbolizing less correlation and lighter means a higher correlation.
#Here there is a low correlation of columns with rain to columns of temperature.
#We can now conclude that the place this was recorded in doesn’t really receive rainfall and is prone to being in colder temperatures.
```

```python
import seaborn as sb
import pandas as pd

dataset = pd.read_csv("assessment_3_dataset_2.csv")
dataset = dataset.rename(columns = {"01/07/2014" : "The date", "76" : "Actual average °F that day", "66" : "Actual minimum °F that day",
                                   "86" : "The actual maximum °F that day", "66.1" : "The average minimum °F of that day since 1901", "85" : "The average max °F that day since 1901",
                                    "48" : "The lowest °F that day since 1901", "97" : "The highest °F that day since 1901", "0.13" : "The actual amount of rain that day", 
                                    "0.16" : "The average amount of rain on that day since 1901", "5.09" : "The highest amount of rain on that day since 1901"})

sb.pairplot(dataset, hue = "The actual maximum °F that day", height = 4) 
#If the height is = 1 then the dots here are really big and not good for gaining info,
#And if it is 9 the names to represent the type of graph is really small to read, so it is better for it to be at 4,
#A challenge I had in implementing this is having to choose the proper amount for height as it'd take time to load an entire pairplot every time
```

```python
X_and_Y_labels = dataset.plot()
X_and_Y_labels.set_xlabel("Amount of days -->")
X_and_Y_labels.set_ylabel("°F degrees")
#The reason I chose this is because we can compare all the columns at once without creating individual visualizations to go back and forth in viewing.
#In this broad time-wise prediction,
#when the data goes 200 days in it visually becomes winter as we can tell by the max and min °F lines being close together.
#Another insight we can gain from this is that at that specific region the chances of rain are really low in comparison since 1901, and that the chances of rain increase when the temperatures go up indicating spring then summer.
#From this, we can guess this place is not likely to be Australia as its average peak temperature is 80 °F, which is 20 °C, and more likely to be within the northern hemisphere.
```
